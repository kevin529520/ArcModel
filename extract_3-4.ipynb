```python
import cv2
import os
from matplotlib import pyplot as plt
import numpy as np

# Define the paths to the cropped image folders
cropped_p3_folder = './cropped_binary_resized_mirror/cropped_binary_resized_mirror_p3/'
cropped_p4_folder = './cropped_binary_resized_mirror/cropped_binary_resized_mirror_p4/'

# Create a folder to save the extracted images
extracted_folder = './p2-7_extracted_cropped_binary/p4_extracted/'
if not os.path.exists(extracted_folder):
    os.makedirs(extracted_folder)

# Get the list of files in cropped_p4 folder
files_p4 = os.listdir(cropped_p4_folder)

# Iterate over the files in cropped_p4 folder
for file_p4 in files_p4:
    # Construct the paths for the corresponding images in cropped_p3 and cropped_p4 folders
    file_p3 = os.path.join(cropped_p3_folder, file_p4.replace('p4','p3'))
    file_p4 = os.path.join(cropped_p4_folder, file_p4)
    
    # Read the images
    img_p3 = cv2.imread(file_p3, cv2.IMREAD_GRAYSCALE)
    img_p4 = cv2.imread(file_p4, cv2.IMREAD_GRAYSCALE)
    

    # plt.imshow(img_p4, cmap='gray')
    # plt.axis('on')  # Turn off axis numbers
    # plt.show()
    # Subtract the images
    extracted_image = cv2.subtract(img_p3, img_p4)
    
# Invert the black and white colors
    # extracted_image = cv2.bitwise_not(extracted_image)

    # 黑色中找到白色的  （查找外轮廓）
    contours, _ = cv2.findContours(extracted_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Iterate over contours
    for contour in contours:
        area = cv2.contourArea(contour)
        if area < 1000:
            cv2.drawContours(extracted_image, [contour], -1, 0, thickness=cv2.FILLED)


    white_pixel_threshold = 5
    for col in range(extracted_image.shape[1]):
        white_pixel_count = np.sum(extracted_image[:, col] == 255)
        if white_pixel_count <= white_pixel_threshold:
            extracted_image[:, col] = 0  # Convert the column to black pixels

# Invert the black and white colors
    extracted_image = cv2.bitwise_not(extracted_image)

    black_pixel_count = np.sum(extracted_image == 0)
    print("Number of black pixels: ", black_pixel_count)
    # Calculate the area of black pixels in square millimeters
    black_area_mm = black_pixel_count * (29/547)**2
    print("Area of black pixels in square millimeters: ", black_area_mm)


    plt.imshow(extracted_image, cmap='gray')
    plt.axis('on')  # Turn off axis numbers
    plt.show()
    # Save the extracted image
    file_p4 = file_p4.replace('p4_','')
    cv2.imwrite(os.path.join(extracted_folder, 'p4_extracted_' + file_p4.split("/")[-1]), extracted_image)
```
