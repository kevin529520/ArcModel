```python
import numpy as np
import csv

def generate_trajectory(x1, y1, z1, x2, y2, z2, x3, y3, z3, x_interval=0.04):
    # 计算x轴上的间隔
    x_values = np.arange(x1, x3, x_interval)
    
    # 对于每个x值，使用线性插值计算对应的y和z值
    y_values = np.interp(x_values, [x1, x2, x3], [y1, y2, y3])
    z_values = np.interp(x_values, [x1, x2, x3], [z1, z2, z3])
    
    return x_values, y_values, z_values

# 示例点的位置   /cm
x1, y2_1, z2_1 = 0, -0.2, 0.7
x2, y2_2, z2_2 = 5, 0.9, 0.85
x3, y2_3, z2_3 = 10, -0.2, 1



# 生成轨迹+
x_traj, y2_traj, z2_traj = generate_trajectory(x1, y2_1, z2_1, x2, y2_2, z2_2, x3, y2_3, z2_3)
y2_pix = 10*(2.8 - y2_traj ) * (580/29)  
z2_pix = 580 - 10*(z2_traj + 0.6) * (580/29)

print(len(y2_traj))
print(len(z2_traj))



# 读取文件
with open('./Data/labels-240301-p1_p6.txt', 'r') as f:
    reader = csv.reader(f)
    data = list(reader)
    for i in range(len(data)):
    # 如果原有的列数不足6列，用空字符串填充
        while len(data[i]) < 6:
            data[i].append('')


for i, (y, z) in enumerate(zip(y2_pix[:200], z2_pix[:200])):
# for i, (y, z) in enumerate(zip(y2_traj[:200], z2_traj[:200])):
#     y = 10 * y  #cm -> mm
#     z = 10 * z   #cm -> mm
    data[i][1] = round(y, 3)
    data[i][2] = round(z, 3)

# 写回文件
with open('./Data/labels-240301-p1_p6.txt', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerows(data)


```

```python
import numpy as np
import csv

def generate_trajectory(x1, y1, z1, x2, y2, z2, x3, y3, z3, x_interval=0.04):
    # 计算x轴上的间隔
    x_values = np.arange(x1, x3, x_interval)
    
    # 对于每个x值，使用线性插值计算对应的y和z值
    y_values = np.interp(x_values, [x1, x2, x3], [y1, y2, y3])
    z_values = np.interp(x_values, [x1, x2, x3], [z1, z2, z3])
    
    return x_values, y_values, z_values
```

```python
# 示例点的位置
x1, y3_1, z3_1 = 0, 0.9, 1
x2, y3_2, z3_2 = 5, -0.2, 1.2
x3, y3_3, z3_3 = 10, 0.9, 1.4



# 生成轨迹+
x_traj, y3_traj, z3_traj = generate_trajectory(x1, y3_1, z3_1, x2, y3_2, z3_2, x3, y3_3, z3_3)
y3_pix = 10*(2.8 - y3_traj) * (580/29)  
z3_pix = 580 - 10*(z3_traj + 0.6) * (580/29)

for x, y, z in zip(x_traj, y3_pix, z3_pix):
    print(f"x: {x:.2f}, y: {y:.8f}, z: {z:.4f}")

# 读取文件
with open('./Data/labels-240301-p1_p6.txt', 'r') as f:
    reader = csv.reader(f)
    data = list(reader)


for i, (y, z) in enumerate(zip(y3_pix[:200], z3_pix[:200])):
# for i, (y, z) in enumerate(zip(y3_traj[:200], z3_traj[:200])):  # 实际焊接位置 非像素
#     y = 10 * y  #cm -> mm
#     z = 10 * z   #cm -> mm
    data[200 + i][1] = round(y, 3)
    data[200 + i][2] = round(z, 3)

# 写回文件
with open('./Data/labels-240301-p1_p6.txt', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerows(data)
```
