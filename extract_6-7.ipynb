```python
import cv2
import os
from matplotlib import pyplot as plt
import numpy as np

# Define the paths to the cropped image folders
cropped_p6_folder = './cropped_binary_resized_mirror/cropped_binary_resized_mirror_p6/'
cropped_p7_folder = './cropped_binary_resized_mirror/cropped_binary_resized_mirror_p7/'

# Create a folder to save the extracted images
extracted_folder = './p2-7_extracted_cropped_binary/p7_extracted/'
if not os.path.exists(extracted_folder):
    os.makedirs(extracted_folder)

# Get the list of files in cropped_p7 folder
files_p7 = os.listdir(cropped_p7_folder)

# Iterate over the files in cropped_p7 folder
for file_p7 in files_p7:
    # Construct the paths for the corresponding images in cropped_p6 and cropped_p7 folders
    file_p6 = os.path.join(cropped_p6_folder, file_p7.replace('p7','p6'))
    file_p7 = os.path.join(cropped_p7_folder, file_p7)
    
    # Read the images
    img_p6 = cv2.imread(file_p6, cv2.IMREAD_GRAYSCALE)
    img_p7 = cv2.imread(file_p7, cv2.IMREAD_GRAYSCALE)
    

    # plt.imshow(img_p7, cmap='gray')
    # plt.axis('on')  # Turn off axis numbers
    # plt.show()
    # Subtract the images
    extracted_image = cv2.subtract(img_p6, img_p7)
    
# Invert the black and white colors
    # extracted_image = cv2.bitwise_not(extracted_image)

    #   
    contours, _ = cv2.findContours(extracted_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Iterate over contours
    for contour in contours:
        area = cv2.contourArea(contour)
        if area < 1000:
            cv2.drawContours(extracted_image, [contour], -1, 0, thickness=cv2.FILLED)


    white_pixel_threshold = 5
    for col in range(extracted_image.shape[1]):
        white_pixel_count = np.sum(extracted_image[:, col] == 255)
        if white_pixel_count <= white_pixel_threshold:
            extracted_image[:, col] = 0  # Convert the column to black pixels

# Invert the black and white colors
    extracted_image = cv2.bitwise_not(extracted_image)

    black_pixel_count = np.sum(extracted_image == 0)
    print("Number of black pixels: ", black_pixel_count)
    # Calculate the area of black pixels in square millimeters
    black_area_mm = black_pixel_count * (29/550)**2
    print("Area of black pixels in square millimeters: ", black_area_mm)


    plt.imshow(extracted_image, cmap='gray')
    plt.axis('on')  # Turn off axis numbers
    plt.show()
    # Save the extracted image
    file_p7 = file_p7.replace('p7_','')
    cv2.imwrite(os.path.join(extracted_folder, 'p7_extracted_' + file_p7.split("/")[-1]), extracted_image)
```
